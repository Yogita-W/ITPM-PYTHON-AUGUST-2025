```python
class Student_FirstName:
    def Std_name(self,Name):
        self.Name = Name
        return f"My Name is {Name}"
class Student_lastName: 
    def Std_lastName(self,lastName):
        self.lastName = lastName
        return f"My  lastName is {lastName}"
       
class Id(Student_FirstName,Student_lastName):
    def Std_id(self,id):
        self.id = id
        return f"{id} This is {self.Name} {self.lastName } id "  
Obj1 = Id()
Obj1.Std_name("Shraddha")
Obj1.Std_lastName("Jadhav")
print(Obj1.Std_id(101))    
```

```python
class GP:
    def Name(self,n1):
        self.n1 = n1
class Parent(GP):
    def Name2(self,n2):
        self.n2 =n2
        return f" I am Parent and my name is {n2} and my father name is {self.n1}"
class Child(Parent) :
    def Name3(self,n3):
        self.n3 =n3
        return f"My name is {n3} {self.n2} and my GrandFather's Name is {self.n1}"    

obj1 = Child()
obj1.Name("Suresh")
obj1.Name2("Rahul Kokate")
print(obj1.Name3("Tanay"))           
```

```python
class Parent:
    def Python_DA(self,P):
        self.P = P
class Child(Parent):
    def Trainer_name(self,n1): 
        self.n1 = n1
        return f"{n1} is {self.P} Trainer"
class Child2(Parent):
    def Trainer2(self,n2):
        self.n2 = n2
        return f"{n2} is {self.P} Trainer"

obj1 = Child()
obj1.Python_DA("Python")
print(obj1.Trainer_name("Yogita"))
obj2 = Child2()
obj2.Python_DA("DA")
print(obj2.Trainer2("Piyush"))   
                  
```
