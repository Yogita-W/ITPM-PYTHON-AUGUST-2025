```python
class Student:
    def __init__(self,Name,Add):
        self.Name = Name
        self.Add = Add 
    def Std_info(self):
        return f'Student Name is {self.Name} and she is from {self.Add}'    
        
        
obj1 = Student('Shrddha','Pune')
obj1.Std_info()
```

```python
class Course:
    def __init__(self,c1,c2='UI'):
        self.c1 = c1
        self.c2 = c2
    def std_c(self):
        return f"{self.c1} , {self.c2}"    
obj3 = Course("python")
obj3.std_c()    
```

```python
class Bank_details:
    def __init__(self,account_no):
        self.account_no = account_no
    def credit(self,amount_credited ):
        self.amount_credited = amount_credited 
        return f'In account number {self.account_no} :Rs {amount_credited} is Credited'
    def debit(self,money):
        self.money = money
        return f'In account number {self.account_no} :Rs {money} is debitted'

Customer = Bank_details('123456789')
print(Customer.credit(10000))
print(Customer.debit(3000)) 
```
