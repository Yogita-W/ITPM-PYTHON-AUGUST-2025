Encapsulation : Binding Data and its realeted methods into single unit.

```python
#Class is example of Encp.
class Math:
    def __init__(self,a,b):
        self.a = a
        self.b = b
        
    def Add(self):
        return f"Addition of {self.a} and {self.b} is : {self.a+self.b}"
    def Sub(self):
        return f"Substraction of {self.a} and {self.b} is : {self.a-self.b}"    
    def Mul(self):
        return f"Multiplication of {self.a} and {self.b} is : {self.a*self.b}" 
    def Div(self):
        return f"Division of {self.a} and {self.b} is : {self.a/self.b}" 
    
obj1 = Math(20,10)
print(obj1.Add())   
print(obj1.Sub()) 
print(obj1.Mul()) 
print(obj1.Div())  
```

```python
class Course:
    def __init__(self, cname, tname):
        self.__cname = cname      # private variable
        self.__tname = tname      # private variable
        self.__duration = None    # will set later using setter

    # Setter method for duration
    def set_duration(self, duration):
        self.__duration = duration

    # Getter method for duration
    def get_duration(self):
        return self.__duration

    # Getter method for course name
    def get_cname(self):
        return self.__cname

    # Getter method for trainer name
    def get_tname(self):
        return self.__tname

    def Coursedetails(self):
        return f"Details of {self.__cname} : trainer name - {self.__tname}, Duration - {self.__duration}"


class Student(Course):
    def Student_info(self, name):
        self.name = name
        # Access private variables using getters
        return f"Name: {self.name}, Course name: {self.get_cname()}, Trainer: {self.get_tname()}, Duration: {self.get_duration()}"


# Create object
Std = Student('UI', 'Sachin Sir')

# Set duration using setter
Std.set_duration('6 months')

# Display details
print(Std.Coursedetails())
print(Std.Student_info('Amol'))
```
