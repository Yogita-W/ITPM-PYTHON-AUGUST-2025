file = open("filename.txt", "mode")
   
   - `mode` can be:
     - `'r'`: Read (default)
     - `'w'`: Write (overwrites if the file exists)
     - `'x'`: Create (fails if the file exists)
     - `'a'`: Append
     - `'b'`: Binary mode
     - `'t'`: Text mode (default)
     - `'r+'`: Read and write
