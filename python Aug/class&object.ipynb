Class and Object:

Object: object is an instance of a class.
Class Syntax: 
              class classname :

```python
class student_info: #class created
    name = 'riya'   #variables  in class
    print(name)
    
obj1 = student_info()   #object created 
```

```python
#declearing variables inside class
class Home :
    hall = 'white'
    kitchen = 'black'
    print(hall + kitchen)

myhome = Home()    
```

```python
#creating class using methods
class cal:
    def add(self,a,b):
        self.a = a
        self.b = b
        return f'Addition of {a} and {b} : {a+b}'

math = cal()
math.add(12,30)     

        
```

```python
#creating multiple objects of one class.
class Students:
    def Python(self,std1,std2):
        self.std1 = std1
        self.std2 = std2
        return f'{std1} and {std2} are from python batch'
obj1 = Students()
print(obj1.Python('Vikram','Pratham'))
obj2 = Students()
print(obj2.Python('Chhaya','Pooja'))
```

```python
#Creating class with multiple function.
class Car :
    def color(self,c):
        self.c = c
        return f'This is {c} color Car'
    def Engin(self,En):
        self.En = En
        return f'Engin of this car is of {En} CC'
    
Obj1 = Car()
print(Obj1.color('blue'))
print(Obj1.Engin(1500))    
```

```python
class VE:
    def Name(self,N1):
        self.N1 = N1
        return f'Name of Person {N1}'
    def Age(self,age):
        self.age = age
        if age>18:
            return f'age of Person is {age} so he/she is Eligible for Voting.....'
        else:
            return f'age of Person is {age} and for Voting age should be greater then 18 so Eligible for Voting.....'
    
Obj2 = VE()
print(Obj2.Name('Karan'))
print(Obj2.Age(22))    
```

```python
class math :
    def add(self,a,b):
        self.a = a
        self.b = b
        return f'Addition of numbers : {a+b}'
    def Sub(self,a,b):
        self.a = a
        self.b = b
        return f'Sub of numbers : {a-b}'
    def Mul(self,a,b):
        self.a = a
        self.b = b
        return f'Mul of numbers : {a*b}'
    
Obj3 = math()
print(Obj3.add(10,20))  
print(Obj3.Sub(20,10))
print(Obj3.Mul(10,4))  
    
    
```

```python
class Math:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def add(self):
        return f"Addition of {self.x} and {self.y} is {self.x+self.y}"
    def sub(self):
        return f"Sub of {self.x} and {self.y} is {self.x-self.y}"

Obj4 = Math(20,30)
print(Obj4.add())
print(Obj4.sub())
    
    
    
```
