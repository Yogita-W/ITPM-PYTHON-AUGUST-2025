```python

menu = {
    "whole_spices": [
        ("jeera", 70),
        ("mustard_seeds", 60),
        ("cloves", 120),
        ("cinnamon", 150),
        ("cardamom", 180),
        ("black_pepper", 160)
    ],
    "powder_spices": [
        ("turmeric_powder", 80),
        ("chili_powder", 90),
        ("coriander_powder", 70),
        ("cumin_powder", 85),
        ("ginger_powder", 100),
        ("garam_masala", 120)
    ],
    "blended_spices": [
        ("sambar_masala", 110),
        ("biryani_masala", 130),
        ("chole_masala", 100),
        ("kitchen_king", 125),
        ("paneer_masala", 115),
        ("pav_bhaji_masala", 105)
    ],
    "dry_items": [
        ("dry_coconut", 90),
        ("kasuri_methi", 70),
        ("tamarind", 60),
        ("dry_ginger", 85),
        ("nutmeg", 95),
        ("black_cardamom", 150)
    ]
}

while True:
    print("====================================")
    print("     Welcome to Indian Spices System")
    print("====================================")
    print("1. Buy Spices")
    print("2. Exit")
    choice = input("Enter your choice (1 for Buy, 2 for Exit): ")

    if choice != "1":
        print("Thank you for visiting Indian Spices Store!")
        break

    order = []

    while True:
        print("\nAvailable Categories:")
        print("1. Whole Spices")
        print("2. Powder Spices")
        print("3. Blended Spices")
        print("4. Dry Items")
        category_choice = input("Enter your choice for category (1-4): ")

        if category_choice == "1":
            category = "whole_spices"
        elif category_choice == "2":
            category = "powder_spices"
        elif category_choice == "3":
            category = "blended_spices"
        elif category_choice == "4":
            category = "dry_items"
        else:
            print("Invalid choice! Please try again.")
            continue

        print(f"\nYou selected {category.replace('_',' ')}:")
        items = menu[category]

        i = 1
        for item, price in items:
            print(i, ".", item, "₹", price)
            i += 1

        item_name = input("Enter the spice name you want to buy (example: jeera): ")

        found = False
        for food, price in items:
            if food == item_name:
                qty = int(input("Enter quantity (in packets): "))
                order.append([food, price, qty])
                print(food, "x", qty, "added to your basket.")
                found = True
                break

        if not found:
            print("Item not found. Please try again.")

        again = input("Do you want to buy more spices? (yes/no): ")
        if again.lower() != "yes":
            break

    if not order:
        print("No items bought. Returning to main menu...")
        continue

    total = 0
    for food, price, qty in order:
        total += price * qty

    name = input("\nEnter your name: ")
    date = input("Enter today's date (DD-MM-YYYY): ")

    # Bill number = length of name * total amount
    bill_number = len(name) * total

    print("\nSelect your payment option:")
    print("1. Cash")
    print("2. UPI")
    print("3. Card")
    pay = input("Enter your choice (1-3): ")

    if pay == "1":
        payment = "Cash"
        print("Please pay cash at the counter.")
    elif pay == "2":
        payment = "UPI"
        upi_id = input("Enter your UPI ID (e.g. name@upi): ")
    elif pay == "3":
        payment = "Card"
        card_no = input("Enter last 4 digits of your card: ")
    else:
        payment = "Invalid Option"

    print("\n ----------- ---Invoice Bill -----------------------")
    
    print("-----------------------------------------------------")
    print("              INDIAN SPICE STORE")
    print("   Near ITP Institute, Python Batch.Pune")
    print("------------------------------------------------------")
    print("Bill No:", bill_number)
    print("Date:", date)
    print("Customer Name:", name)
    print("------------------------------------------------------")
    print("Item                 Qty      Price      Total")
    print("------------------------------------------------------")

    for food, price, qty in order:
        print(f"{food:20}{qty:5}{price:8}{price*qty:10}")

    print("------------------------------------------------------")
    print(f"{'Total Amount:':30} ₹{total}")
    print("------------------------------------------------------")
    print("Payment Mode:", payment)
    if pay == "2":
        print("UPI ID:", upi_id)
    elif pay == "3":
        print("Card No: ****", card_no)
    print("------------------------------------------------------")
    print("      Thank you for visiting Indian Spices Store!")
    print("              Have a flavorful day! ")
    print("--------------------------------------------------")

    again = input("\nDo you want to buy again? (yes/no): ")
    if again != "yes":
        print("      Thank you for visiting! Come again soon!   ")
        break
```
