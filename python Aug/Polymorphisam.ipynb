```python
class ITP:
    def owner(self, oname):
        self.oname = oname

class Manager(ITP):
    def managername(self,m1name):
        self.m1name = m1name
        return f"I {m1name} am from first child class and the owner is {self.oname}."
class Manager2(ITP):
    def manager2name(self,m2name):
        self.m2name=m2name
        return f"I {m2name} am from second child class."
    
class HR(Manager, Manager2):
    def HR(self,hr):
        self.hr = hr
        return f"I {hr} am using multiple inheritance."
    
obj = HR()
print(obj.HR("Rohit"))
print(obj.managername("Richa"))
print(obj.manager2name("Vedant"))
print(obj.owner("Owner"))
```
