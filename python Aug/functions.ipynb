```python
def Even_add(l1): #Create function with l1 seq..
    l2 = []       #Empty list for storing all even numbers.
    for i in l1:  # to perform itration on l1
        if i%2==0: # cond for checking i is even or not
            l2.append(i) #
    return sum(l2)        
Even_add([12,32,43,2,3,32,21])
#even:12,32,2,32
```

```python
def Grade_sys(marks):
    if marks>90 and marks<101: 
        return f"As you have {marks} so have A grade" 
    elif marks>80 and marks<91:
        return f"As you have {marks} so have B grade" 
    elif marks>60 and marks<81:
        return f"As you have {marks} so have C grade" 
    else :
        return f"As you have {marks} so have F grade"
    

print(Grade_sys(80))
print(Grade_sys(95))
print(Grade_sys(43))
print(Grade_sys(66))
print(Grade_sys(12))
```
